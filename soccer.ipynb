```python
#EN: We import the Pandas Library so that we can use the features that exists in Pandas Library. And then we make an alias named 'pd'
#ID: Kita pertama import Pandas Library supaya bisa pakai segala fitur yang ada di Pandas. Kemudian kita buat alias dengan nama 'pd'
import pandas as pd

#EN: pd.read_csv is the methods to ask Pandas to read the file that have .csv extension 
#ID: pd.read_csv artinya kita meminta Pandas untuk membaca file dengan ekstensi .csv
data_frame = pd.read_csv('data.csv')
# ========================================================
#EN: Now we start to exploring the Data
#ID: Sekarang kita mulai eksplorasi Data yang ada
# ========================================================
#EN: data_frame.shape has the meaning to prints number of rows and columns in dataframe
#ID: data_frame.shape artinya meminta Pandas untuk memberitahu berapa total baris dan kolom pada data 
data_frame.shape
```

```python
#EN: data_frame.describe() means that we summary statistics for numerical columns
#ID: data_frame.describe() memiliki makna bahwa kita meminta Pandas buat Summary statistik untuk kolom Numerik
data_frame.describe()
```
